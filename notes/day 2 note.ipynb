## List/dict comprehensions

The idea is that i can use comprehensions to make the code simpler and easier to read


```python
doubles = [] 
for x in range(1,11):
    doubles.append(x * 2)

print(doubles, x)
#To do the same with comprehensions i can:
#doubles = [expression for value in iterable if condition]

doubles = [x * 2 for x in range(1,11) if 1]

print(doubles, x)
print(type(doubles))
```

```python
fruits = ["apple", "banana", "watermelon"]

fruits = [whatever.upper() for whatever in fruits] # whatever showcases the use of .upper() etc in this case and in general

print(fruits)
print(type(fruits))
```

2. Functions — arguments, default values, *args, **kwargs, return values

The goal here to revise how i make functions and how i use them properly

```python
def function(argument1, argument2): # No need to call what is the type of each argument while defining the function
    return argument1, argument2

def function_2(argument1 = 0, argument2 = 0): # This is called default argument. If i call function_2(1,2) ofc it will be overwritten
    return argument1 * argument2

print(function_2())
print(function_2(1,2))

# def function_3(arg1 = 0, arg2): # Thats a bad use the default arguments should be last in order like (arg2, arg1 = 0)
```

```python
# When it comes to the types of the args i have to keep in mind that:
# If arg is Mutable ->  it changes the original arg - list, dict, set, and most custom objects
# If arg is Immutable -> it doesnt change the original arg - int, float, bool, str, tuple, frozenset

# Remember i can check what type it is thru:
x = "string"
print(id(x))

x += "addition"
print(id(x))

# Change in id means its Immutable
```

```python
# Using a mutable as default value will have "memory" of all the times it gets called

def function(item, box=[]):
    box.append(item)
    return box

print(function(1))
print(function(1))
print(function(1))

# The proper way to fix it is: (If it needs fixing maybe its intended :P)
def function2(item, box=None):
    if box is None: # Everytime it gets no box it creates a new one
        box = []
    box.append(item)
    return box

print(function2(1))
print(function2(1))
print(function2(1))
```
