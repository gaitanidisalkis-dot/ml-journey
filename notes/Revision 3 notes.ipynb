# Single-purpose functions

It is a nice tactic to make single-purpose functions instead of multi-purpose functions. 

The orchistration will have more functions to use in the first case but it will be easy to read almost like normal text.

````python
# Bad - Multi-purpose function
def process(filename):
    with open(filename) as f:
        lines = f.readlines()
    numbers = [float(line) for line in lines]
    avg = sum(numbers) / len(numbers)
    print(f"Average: {avg}")

# Good - Single-purpose function
def read_numbers(filename):
    with open(filename) as f:
        return [float(line) for line in f]

def average(numbers):
    return sum(numbers) / len(numbers)

# Orchestration of multi-purpose
# numbers = process("data.txt")

# Orchestration of sigle-purpose
# numbers = read_numbers("data.txt")
# print(f"Average: {average(numbers)}")

````

# DRY — Don't Repeat Yourself

It is actually the idea that if you need to write the same lines of code multiple times its better to make a function for it.

Most of the times we follow "three strikes and you refactor".
