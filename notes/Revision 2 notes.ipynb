## List/dict comprehensions

The idea is that i can use comprehensions to make the code simpler and easier to read


```python
doubles = [] 
for x in range(1,11):
    doubles.append(x * 2)

print(doubles, x)
#To do the same with comprehensions i can:
#doubles = [expression for value in iterable if condition]

doubles = [x * 2 for x in range(1,11) if 1]

print(doubles, x)
print(type(doubles))
```

```python
fruits = ["apple", "banana", "watermelon"]

fruits = [whatever.upper() for whatever in fruits] # whatever showcases the use of .upper() etc in this case and in general

print(fruits)
print(type(fruits))
```

```python
# Using a mutable as default value will have "memory" of all the times it gets called

def function(item, box=[]):
    box.append(item)
    return box

print(function(1))
print(function(1))
print(function(1))

# The proper way to fix it: (If it needs fixing maybe its intended :P)
def function2(item, box=None):
    if box is None: # Everytime it gets no box it creates a new one
        box = []
    box.append(item)
    return box

print(function2(1))
print(function2(1))
print(function2(1))
```

There are ways to pass multiple args to a function without having to name each and every one of them beforehand.

*args = allows to pass multiple non-key arguments . 
1) It is a tuple of values. -> funct(*args) -> *args = (1,2,3)
2) I can actually use anyword with * like *nums


**kwargs = allows to pass multiple keyword - arguments 
1) It is a dict like x = {"name": "Anna", "age": 30}  BUT i write is as function(name = "Anna", age = "30")
2) I can also use any word instead of kwargs like **andress

```python
def function(*args):
    total = 0
    for num in args:
        total += num
    return total

print(function(1,2,3,2))
```

```python
def functionk(**kwargs):
    print(f"{kwargs.get('age')}") # prints 100 as in age

print(functionk(name2 = "Alkis", age ="100")) # prints None because it doesnt return a thing
```

# Scope — local vs global

global variables are visible to every function created but local ones only to the function that created them. Pretty basic stuff


```python
test_variable = "global"
test_num = 1

def test_global():
    test_variable = "local"
    print (test_variable)
    print (test_num)

test_global()
print (test_variable)
```
